```python
import pandas as pd
import numpy as np

from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
# Load data into a dataframe.
pow_output = pd.read_csv('Cleaned PV System df.csv')
```

```python
# Convert temperature from temperature_2_m_above_gnd to Kelvin.
pow_output['temperature_2_m_above_gnd'] += 273.15
```

```python
# Feature engineering includes taking natural log transformation of the numeric variables.
for col in pow_output.columns:
    if col != 'power_output':
        pow_output.loc['log_' + col] = np.log(pow_output[col])
```

```python
# Feature Selection entails removing collinear features above the specified correlation coefficient of 0.6.
```

```python
no_coll_feat = pow_output.drop(['total_cloud_cover_sfc', 'high_cloud_cover_high_cld_lay', 'medium_cloud_cover_mid_cld_lay', 'low_cloud_cover_low_cld_lay', 'wind_speed_10_m_above_gnd', 'wind_direction_10_m_above_gnd', 'wind_speed_80_m_above_gnd', 'wind_direction_80_m_above_gnd', 'wind_direction_80_m_above_gnd', 'wind_speed_900_mb', 'wind_gust_10_m_above_gnd', 'wind_direction_900_mb'], axis=1)
```
