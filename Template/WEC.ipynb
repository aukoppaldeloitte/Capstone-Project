```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# created dataframe (WEC) with World Energy Consumption csv file
WEC = pd.read_csv("World Energy Consumption.csv")
```

```python
WEC
```

```python
# Created dataframe (ECC) for renewable/non-renewable energies (wind, gas, nuclear, solar, oil, hydro & coal) in major consuming countries (China, India & United States) from 2000 - 2020.
```

```python
energy_types_conditions_df_1 = WEC["country"].isin(["China", "India", "United States"])
energy_types_conditions_df_2 = WEC["year"] > 2000
energy_types_conditions_df_3 = WEC["year"] < 2020
```

```python
ECC = WEC[energy_types_conditions_df_1 & energy_types_conditions_df_2 & energy_types_conditions_df_3][["year","country","wind_consumption","gas_consumption","nuclear_consumption","solar_consumption","oil_consumption","coal_consumption","hydro_consumption"]]
```

```python
ECC
```

```python
ECC.dtypes
```

```python
# Visualize the energy consumption (measured in terawatt-hours) for each of the countries.
```

```python
# Energy consumption for China.
Energy_consumption_for_china = ECC["country"].isin(["China"])
China = ECC[Energy_consumption_for_china]
China
```

```python
# China major renewable energy consumption types in terawatt-hours.
plt.plot(China["year"], China["wind_consumption"], label = 'Wind')
plt.plot(China["year"], China["solar_consumption"], label = 'Solar')
plt.plot(China["year"], China["hydro_consumption"], label = 'Hydro')
plt.title('China Renewable Energy Consumption 21st Century')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Terawatt Hours (TWh)', fontsize=15)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# China major non-renewable energy consumption types in terawatt-hours.
plt.plot(China["year"], China["gas_consumption"], label = 'Gas')
plt.plot(China["year"], China["oil_consumption"], label = 'Oil')
plt.plot(China["year"], China["coal_consumption"], label = 'Coal')
plt.plot(China["year"], China["nuclear_consumption"], label = 'Nuclear')
plt.title('China Non-renewable Energy Consumption 21st Century')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Terawatt Hours (TWh)', fontsize=15)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# Model implementations for non-renewable energies.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier
```

```python
X = ECC[["gas_consumption", "oil_consumption", "coal_consumption", "nuclear_consumption"]]
Y = ECC[["gas_consumption"]]
X_train, X_test, y_train, y_test = train_test_split(X,Y)
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
